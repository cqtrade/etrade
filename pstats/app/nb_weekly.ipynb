```python
data = vbt.BinanceData.download(
    symbols = symbols,
    start=start,
    end=end,
    interval=interval,
    missing_index='drop'
)
price = data.get('Close')
closes = data.get('Close')
highs = data.get('High')
lows = data.get('Low')
volumes = data.get('Volume')

# windows = np.arange(2, 101)
# fast_ma, slow_ma = vbt.MA.run_combs(price, window=windows, r=2, short_names=['fast', 'slow'])
# entries = fast_ma.ma_crossed_above(slow_ma)
# exits = fast_ma.ma_crossed_below(slow_ma)

# pf_kwargs = dict(size=np.inf, fees=0.001, freq='1D')
# pf = vbt.Portfolio.from_signals(price, entries, exits, **pf_kwargs)

# fig = pf.total_return().vbt.heatmap(
#     x_level='fast_window', y_level='slow_window', slider_level='symbol', symmetric=True,
#     trace_kwargs=dict(colorbar=dict(title='Total return', tickformat='%')))
# fig.show()
```

```python
windows = np.arange(5, 101, 5)
fast_ma, slow_ma = vbt.MA.run_combs(
    price, window=windows, r=2, short_names=['fast', 'slow'])
entries = fast_ma.ma_crossed_above(slow_ma)
# exits = fast_ma.ma_crossed_below(slow_ma)

print(type(entries))

print(entries)

# entries = fast_ma.ma_crossed_below(price) & fast_ma.ma_crossed_above(slow_ma)
exits = fast_ma.ma_crossed_above(price)
vbt.Portfolio.from_orders(price, entries, exits,
                          init_cash=1000, size=np.inf, fees=0.001)

pf = vbt.Portfolio.from_signals(price, entries, exits,
                                init_cash=1000,
                                size=np.inf,
                                sl_stop=0.1,
                                fees=0.001,)

figr = pf.total_return().vbt.heatmap(
    x_level='fast_window', y_level='slow_window', symmetric=True,
    trace_kwargs=dict(colorbar=dict(title='Total return', tickformat='%')))
figr.show()

fige = pf.trades.expectancy().vbt.heatmap(
    x_level='fast_window', y_level='slow_window', symmetric=True,
    trace_kwargs=dict(colorbar=dict(title='Expectancy')))

fige.show()

figw = pf.trades.win_rate().vbt.heatmap(
    x_level='fast_window', y_level='slow_window', symmetric=True,
    trace_kwargs=dict(colorbar=dict(title='Win Rate', tickformat='%')))

figw.show()

figc = pf.trades.count().vbt.heatmap(
    x_level='fast_window', y_level='slow_window', symmetric=True,
    trace_kwargs=dict(colorbar=dict(title='Trade count')))

figc.show()
```
